```python
import numpy as np
import pandas as pd
from sklearn import linear_model

df = pd.read_csv('/content/drive/MyDrive/advertising.csv')
df
```

```python
df.Newspaper.mean()
df.Newspaper.fillna(df.Newspaper.mean(), inplace = True)
df
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[["TV", "Radio", "Newspaper"]],df.Sales)
lr.predict([[200, 20, 35.2]]) # 17.66773699
lr.coef_ # 0.05444333, 0.1068452 , 0.00064038 represents m1, m2, m3 in the formula
lr.intercept_ # 4.619625400823253 represents b in the formula
# y= m1x1 + m2x2 + m3x3 + b (where y represents predicted_sales)
predicted_sales = 0.05444333*200 + 0.1068452*20 + 0.00064038*35.2 + 4.619625400823253
print (predicted_sales)
```
